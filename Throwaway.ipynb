```python
import pandas as pd

csv = pd.read_csv('201904 sales reciepts.csv')
df = pd.DataFrame(csv)

df = df[df['product_id'].between(1, 21)]
quantity_sum_by_product_id = df.groupby('product_id')['quantity'].sum()

quantity_sum_by_product_id_sorted = quantity_sum_by_product_id.sort_values(ascending=False)

print(quantity_sum_by_product_id_sorted)
```

```python
filtered_df = df[df['product_id'].isin([6, 1, 8])]
final_out = filtered_df.groupby(['transaction_date', 'product_id'])['quantity'].sum()
final_out=final_out.reset_index().rename(columns={'quantity': 'quantity'})
final_out
# final_out.to_csv('Top3TeaCoffee.csv')
```

```python


date_range = pd.date_range(start='2019-04-01', end='2019-04-30', freq='D')  
product_ids = [1, 6, 8]
all_combinations = pd.DataFrame([(date, product_id) for date in date_range for product_id in product_ids], columns=['transaction_date', 'product_id'])


all_combinations['transaction_date'] = pd.to_datetime(all_combinations['transaction_date'])
final_out['transaction_date'] = pd.to_datetime(final_out['transaction_date'])

merged_data = pd.merge(all_combinations, final_out, on=['transaction_date', 'product_id'], how='left')


merged_data['quantity'].fillna(0, inplace=True)
merged_data

merged_data.to_csv('Top3TeaCoffee.csv', index=False)
```
